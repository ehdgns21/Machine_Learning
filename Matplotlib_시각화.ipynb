```python
import numpy as np
import matplotlib.pyplot as plt # pyplot 함수 불러와서 plt로 앨리어스
%matplotlib inline
```

```python
plt.figure
plt.plot([1,2,3,4,5,6,7,8,9,8,7,6,5,4,3,2,1,0])
plt.show()
```

```python
t = np.arange(0,12,0.01)
y = np.sin(t)
```

```python
plt.figure(figsize=(10,6))
plt.plot(t,y)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, y) # 사용할 데이터 세팅
plt.grid() # 그리드 그리기

plt.xlabel('time') #x축 라벨 붙이기
plt.ylabel('Amplitude') #y축 라벨 붙이기
plt.title('Example of sinewave') # 그래프 제목 붙이기

plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(t, np.sin(t))
plt.plot(t, np.cos(t)) # cos 그래프 추가

plt.grid()  #그래프에 격자무늬
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.title('Example of sinewave')
plt.show()
```
